```python
origin_a = ["https://demo.websec.saarland:443"] * 6
origin_b = ["https://demo.websec.saarland:443", "https://demo.websec.saarland",
            "http://demo.websec.saarland:443", "https://demo.websec.saarland:80",
            "https://not.websec.saarland:443","https://secweb.saarland:443"]
origin = ["Same-Origin", "Same-Origin", "Cross-Origin", "Cross-Origin", "Cross-Orgin", "Cross-Origin"]
site = ["Same-Site", "Same-Site", "Same-Site", "Same-Site", "Same-Site", "Cross-Site"]
exp = ["Exact match", "HTTPS default port is 443", "Different scheme", "Different port", "Different subdomain", "Different site"]

df = pd.DataFrame({"Origin A": origin_a, "Origin B": origin_b, "Origin?": origin, "Site?": site, "Explanation": exp}).set_index(["Origin A", "Origin B"])  
display(df)
with open("tables/origin.tex", "w") as f:
    f.write(df.to_latex())
```
